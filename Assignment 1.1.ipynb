```python
# -*- coding: utf-8 -*-
"""
Created on Tue Feb 20 16:31:02 2018

@author: zabiulla.khan
"""
#integrers
x = -10
type(x)
print(x)

#float
x=-1.01
type(x)
print(x)

#boolean
b1 =True
b2 = False
print(b1)
print(b2)

#strings
x = 'hi'
y = "/hello"
z = '''Good                                                                                                                                                                                                                                                                                                                                                                      
Bad'''
print(x)
print(y)
print(z)

# complex numbers: note the use of `j` to specify the imaginary part
x = 1.0 - 1.0j
type(x)
print(x)
print(x.real,x.imag)

#variables
ten = 10
ten
print(ten)

#dynamic typing
ten = 'ten'
print(ten)

#Strong typing
#'Day' + 1
x = 'Day ' + str(1)
print(x)

#Simple Expressions
True and False
True or False
not True
not False
True is True
True is False
'a' is 'a'

#Branching (if(elif/else))
i=3
if i<3:
    print('less than 3')
elif i<5:
    print('less than 5')
else:
    print('5 or more' )
    
#Lists
l1 = []
print(l1)
l2 = list()
print(l2)
l3 = ['a','b','c']
print(l3)
l4 = ['a','c']
l5 = list(l4)
print(l5)
l4.insert(1,'b')
print(l4)
l6 = list('abcdef')
print(l6)

#Iterating
for i in l6:
    print(i)
    
#Program to find sum of all numbers in a list
    
#List of numbers
l1 = [12,10,6,8,1,98,1,-1]

#variable to store the sum
sum = 0

#Iterate over the list
for i in l1:
    sum = sum + i
    
#Display the sum
print(sum)

# for loop with else
for i in l1:
    print(i)
else:
    print('no items left')

#Usage of while loop
#Program to add natural numbers upto n
# Sum of 1+2+...n

#Take input from the user
n = int(input("Enter n:"))

# Initialize sum and counter
sum = 0
i = 0

while i <= n:
    sum = sum + i
    i = i + 1  
    
# print the sum
print ("The sum is ", sum)
    
# range() function
print(range(10))
print(range(1,7))
print(list(range(1,7)))
print(range(0, 15, 3))
print(list(range(0, 15, 3)))

# Program to iterate through a list using indexing
food = ['fruits', 'vegetables', 'non-veg']
# iterate over the list using index
for i in range(len(food)):
    print("I like", food[i])
    
# Use of break statement inside loop
for i in "String":
    if (i=="i"):
        break
    print(i)
print("End")

# Use of continue statement inside loop
for i in "String":
    if (i=="i"):
        continue
    print(i)
print("End")

#Program to get input from user
name = input("what is your name?")
print("It was nice talking you " + name + "!")
age = input("what is your age?")
print("Hey, you are already " + age + " years old " + name + "!" )

```
